## Combinando Dataframes

### Concatenar

```python
#creando dataframes

df1=pd.DataFrame({'A':['A0','A1','A2','A3','A4'],
 'B':['B0','B1','B2','B3','B4'],
 'C':['C0','C1','C2','C3','C4'],
 'D':['D0','D1','D2','D3','D4'],
 })

df2=pd.DataFrame({'A':['A5','A6','A7','A8','A9'],
 'B':['B5','B6','B7','B8','B9'],
 'C':['C5','C6','C7','C8','C9'],
 'D':['D5','D6','D7','D8','D9'],
 })
```

```python
#Concatenar dos dataframes
pd.concat([df1,df2]) #POR DEFAULT CONCATENA POR FILAS
#Los indices heredan su valor original
#Para corregir o normalizar/reorganizar los indices se puede agregar lo siguiente
pd.concat([df1,df2], ignore_index=True) #Reordena los indices del nuevo dataframe
```

```python
pd.concat([df1,df2], axis=1) #Concatena por columnas
#Es muy raro usar esto, normalmente se realiza por filas
```

### Merge

```python
izq= pd.DataFrame({'Key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3'],
 })

der= pd.DataFrame({'Key':['k0','k1','k2','k3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3'],
 })
```

```python
izq
```

```python
der
```

## Pivot y Melt

### Pivot_Table
Esta función puede traer recuerdos a las personas interesadas en el mundo del SQL, ya que Oracle, PostgreSQL y otros motores de bases de datos la tienen implementada desde hace muchos años. Pivot, básicamente, transforma los valores de determinadas columnas o filas en los índices de un nuevo DataFrame, y la intersección de estos es el valor resultante.


```python
df_books.pivot_table(index='Author',columns='Genre',values='User Rating')
```

Como resultado, los valores de Author pasan a formar el índice por fila y los valores de Genre pasan a formar parte de los índices por columna, y el User Rating se mantiene como valor.

image.png

Por supuesto, para este caso, un Author suele tener un solo género literario, así que no es una transformación muy útil, pero veamos si podemos lograr algo mejor.

```python
df_books.pivot_table(index='Genre',columns='Year', values='User Rating',aggfunc='sum')
```

En este caso tenemos por cada género, la suma a lo largo de los años. Esto es mucho más interesante, ¿verdad? La mejor noticia es que no solo podemos obtener la suma, también podemos obtener la media, la desviación estándar, el conteo, la varianza, etc. Únicamente con cambiar el parámetro aggfunc que traduce función de agrupamiento.

### Melt
El método melt toma las columnas del DataFrame y las pasa a filas, con dos nuevas columnas para especificar la antigua columna y el valor que traía.

Por ejemplo, simplemente al imprimir las cinco primeras filas del DataFrame con las columnas de Name y Genre se tiene este resultado.

```python
df_books[['Name','Genre']].head(5)
```

```python
#Aplicando Melt
df_books[['Name','Genre']].head(5).melt()
```

```python
#Otra forma de aplicar Melt
df_books.melt(id_vars='Year',value_vars='Genre')
```

Simplemente, podemos seleccionar las columnas que no quiero hacer melt usando el parámetro id_vars. Para este caso Year y también la única columna que quiero aplicar el melt, para este caso Genre con la propiedad value_vars.

## Apply

Nos ayuda a aplicar funciones a nuestro Dataframe, definidas previamente en Python

```python
#Defino una funcion en Python
def two_times (value):
  return value*2
```

```python
#Aplicando la función
df_books['User Rating'].apply(two_times)
```

```python
#Puedo agregarle una nueva columna al Dataframe con ese valor que estamos calculando con la función creada
df_books['Rating_2'] = df_books['User Rating'].apply(two_times)
df_books
```

Esto es muy poderoso para aplicar cualquier tipo de función directo al dataframe

```python
#Puedo incluso utilizar funciones de tipo Lambda sin tener que definir funciones en memoria
df_books['Rating_2'] = df_books['User Rating'].apply(lambda x : x * 3)
df_books
```

```python
#Usar distintas columnas para las operaciones
df_books.apply(lambda x : x['User Rating'] * 10 if x['Genre'] == 'Fiction' else x['User Rating'], axis=1)
#Debo especificar que es axis=1 (columnas) para que reconozca 'Genre' de lo contrario lanza error
```

Estoy indicando que multiplique por 10 el User Rating si y sólo si se cumple la condición, de lo contrario (else) que lo deje igual al original 'User Rating'. Esto me ahorra escribir código y es mucho más óptimo (rápido) que usar un for o algún otro Loop. Es recomendable entonces usar Apply para aplicar funciones directo en el dataframe.
